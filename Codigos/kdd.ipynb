#### Gerar matriz de correlação (de Pearson) de variáveis
Lembrando que 1 é a correlação direta perfeita, -1 é a correlação negativa perfeita e valores próximos de 0 indicam a falta de exixtência ou dificuldade de estabelecer correlação

Por alguma razão não está funcionando.

```python
dados.corr()
```

```python
mask = np.triu(np.ones_like(dados.corr(), dtype=np.bool_))
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(dados.corr(), mask=mask, square = True, annot = True, vmin = 1, vmax = 1)
plt.show()
```
