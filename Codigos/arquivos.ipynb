## Arquivos

Usando API que retorna um arquivo Json

```python
import csv 
import pandas as pd
```

```python
arquivo = open('machado de assis.txt', 'r')
```

ler linhas e fechar o arquivo e mostrar o conteúdo 

```python
conteudo = arquivo.readlines()
arquivo.close()
```

```python
print(conteudo)
for linha in conteudo:
    print(linha)
```

alterar texto

```python
x = ''

for linha in conteudo:
    if linha!='':
        x+=linha

print(x)
x = x.replace('Rio de Janeiro', 'Ribeirão Preto')
print('-------------')
print(x)
```

salvar em outro arquivo

```python
arq_new = open('machado_de_assis_2.txt', 'w')
arq_new.writelines(x)
arq_new.close()
```

```python
import csv
with open('Tesla.csv', 'r') as f:
    leitor = csv.reader(f, delimiter=',')
    for linha in leitor:
        data = linha[0]
        fechamento = linha[4]
        volume = linha[5]
        print('Data:', data, '--Fechamento: ', fechamento, '--volume', volume)
```
